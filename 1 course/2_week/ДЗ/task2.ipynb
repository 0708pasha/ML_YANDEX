```python
import numpy as np
import math
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBacket.figure_format = 'svg'

from pylab import rcParams
rcParams['figure.figsize'] = 8, 5
```

```python
def f(x):
    return np.sin(x/5.0)*np.exp(x/10.0)+5*np.exp(-x/2.0)
```

```python
f(0)
```

```python
A2 = np.array([[1.,1.],[1.,15.]])
b2 = np.array([f(1.),f(15.)])
r2 = np.linalg.solve(A2,b2)
r2
```

```python
def f2(x):
    return r2[0]+r2[1]*x
```

```python
x = np.linspace(0.,16.,100)
```

```python
A3 = np.array([[1.,1.,1.],[1.,8.,64.],[1.,15.,225.]])
b3 = np.array([f(1.),f(8.),f(15.)])
r3 = np.linalg.solve(A3,b3)
r3
```

```python
def f3(x):
    return r3[0]+r3[1]*x+r3[2]*x**2
```

```python
A4 = np.array([[1.,1.,1.,1.],[1.,4.,16.,64.],[1.,10.,100.,1000.],[1.,15.,225.,15**3]])
b4 = np.array([f(1.),f(4.),f(10.),f(15.)])
r4 = np.round(np.linalg.solve(A4,b4),2)
r4
```

```python
r4 = np.linalg.solve(A4,b4)
r4
```

```python
def f4(x):
    return r4[0]+r4[1]*x+r4[2]*x**2+r4[3]*x**3
```

```python
plt.plot(x,f(x), label = 'f(x)')
plt.plot(x,f2(x), label = 'f2(x)')
plt.plot(x,f3(x), label = 'f3(x)')
plt.plot(x,f4(x), label = 'f4(x)')
plt.legend()
```
