```python
import numpy as np
import math
from scipy.optimize import minimize
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x/5.0)*np.exp(x/10.0)+5*np.exp(-x/2.0)
```

```python
fig = plt.subplots()
x = np.linspace(0, 30)
plt.plot(x,f(x), label = 'f(x)')
plt.legend()
```

```python
scipy.optimize.minimize (f,30) 
```

```python
res1 = scipy.optimize.minimize (f,2, method='BFGS') #градиентны методы оптимизации
res1
```

```python
res2 = scipy.optimize.minimize (f,30, method='BFGS') #градиентный метод оптимизации
res2
```

```python
a = str(round(res1.fun,2))
b = str(round(res2.fun,2))
```

```python
with open('task1.txt', 'w') as out:
    out.write(str(a)+' '+str(b))
```
