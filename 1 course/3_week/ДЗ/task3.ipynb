```python
import numpy as np
import math
from scipy.optimize import minimize
from scipy.optimize import differential_evolution
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x/5.0)*np.exp(x/10.0)+5*np.exp(-x/2.0)
```

```python
def fint(x):
    return f(x).astype(int)
```

```python
f(1),fint(1)
```

```python
fig = plt.figure(figsize=(10,6)).subplots()
x = np.linspace(0, 30)
plt.plot(x,f(x), label = 'f(x)')
plt.plot(x,fint(x), label = 'fint(x)')
plt.legend()
```

```python
res1 = minimize (fint,30, method='BFGS') #градиентный метод оптимизации
res2 = differential_evolution (fint, [(1, 30)])
res1, res2
```

```python
a = str(res1.fun[0])
b = str(res2.fun)
```

```python
with open('task3.txt', 'w') as out:
    out.write(a+' '+b)
```
