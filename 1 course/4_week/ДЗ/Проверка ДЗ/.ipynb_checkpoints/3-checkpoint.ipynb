Возьмем мое любимое непрерывное распределение - хи-квадрат или распределение Пирсона. (См. Wiki: https://go-gl.com/xKMtstPae).

$\chi^2$ - это распределение суммы квадратов независимых стандартных нормальных случайных величин.

$\sum\limits_{i=1}^k X^2_i = \chi^2(k)$, где $X_i = (i=1,2,...,k)$ независимые нормальные случайные величины, а k - число степеней свободы. 

```python
# Импорт необходимых библиотек
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Возьмем для примера число степеней свободы k=4.
# Присвоим переменной значение функциии хи-квадрат с k=4
chi2_rv = sts.chi2 (4)
# Переменной sample присвоим массив значений выборки объема 1000, сгенерированной из распределения хи-квадрат 
sample = chi2_rv.rvs(1000)
```

```python
# значения по оси абсцисс
x = np.linspace(-1,20,100)
# переменная pdf содержит теоритеческую плотность распределения нашей случайной величины
pdf = chi2_rv.pdf(x)
```

```python
plt.plot(x, pdf, label='Theoretical PDF')
plt.hist(sample, density = True, label='histogram')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper right')
```

# Оценка распределения выборочных средних

Для распределения $\chi^2$:

математическое ожидание: $k$

дисперсия: $2k$

Значения параметров нормального распределения с помощью которых будем приближать распределение выборочных средних:

$\mu = \mathbb{E}X$

$\sigma^2 = \mathbb{D}[X]$

Для нашего случая математическое ожидание:

$\mu = k$

стандартное отклонение на основании выборочной дисперсии:

$\sigma = \sqrt{\frac{\mathbb{D}[X]}{n}} = \sqrt{\frac {2k}{n}}$


Cгенерируем 1000 выборок случайных величин из нашего распределения объемом n=5 n=10 n=50.

```python
# Запишем все вычисления в функцию clt, принимающую в качестве параметра объем выборки - n

def clt(n):
    mean_array = [] # пустой массив для записи средних значений выборок

    # генерируем 1000 выборок объемом n
    for i in range(1000):
        sample = chi2_rv.rvs(n) # генерация случайной выборки объемом n
        sample_mean = np.mean(sample) # вычисление среднего значения
        mean_array.append(sample_mean) # добавление среднего в массив

    # выводим гистограмму выборочных средних
    plt.hist(mean_array, normed=True, label='means histogram') 

    # Вычисляем значения параметров нормального распределения (используем теоретические среднюю и дисперсию)
    mu = 4. # математическое ожидание
    sigma = (4.*2./n) ** 0.5 # сигма
    norm = sts.norm(mu, sigma)
    
    # Выводим график теоретической плотности нормального распределения
    x = np.linspace(0,15,100)
    pdf = norm.pdf(x)
    plt.plot(x, pdf, linewidth = 2.5, label='Theoretical PDF')
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
    plt.legend(loc='upper right')
    plt.title("Mu=" + str(mu)+" "+"Sigma="+str(sigma))
```

```python
# Вызываем нашу функцию с n = 5
clt (5)
```

```python
# Вызываем нашу функцию с n = 10
clt (10)
```

```python
# Вызываем нашу функцию с n = 50
clt (50)
```

Вывод: при увеличении объема выборки увеличивается точность аппроксимации графика плотности нормального распределения и гистограмм выборочных средних.  При увеличении объема выборки уменьшается дисперсия. Графики сходятся к 4 (выбранному нами числу степеней свободы).
