```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

# Любимое распределение

```python
my_rv = sts.expon(loc=0.0, scale=1.2)
mnogo = 1000
print("Теоретическое среднее %f\nДисперсия %f\nСтандартное отклонение %f" % (my_rv.mean(), my_rv.var(), my_rv.std()))
```

```python
# Сделаем выборку значений и построим гистограмму распределения значений в выборке
sample = my_rv.rvs(mnogo)
plt.hist(sample, bins=30, normed=True)

# на том же графике построим теоретическую плотность распределения:
x = np.linspace(0, 9, 100)
pdf = my_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
None
```

## Выборки по 5 

```python
# делаем много (1000) выборок по 5
samples_n5 = [my_rv.rvs(size=5) for i in range(mnogo)]

# Средние значения выборок
avgs_for_5 = [sum(rv)/len(rv) for rv in samples_n5]

# Нормальное распредедение для средних значений выборок по 5
norm_for_5 = sts.norm(loc=my_rv.mean(), scale=(my_rv.var()/5)**0.5)

print("""Параметры для нормального распределения средних значений случайной выборки 5 Среднее значение %f\nДисперсия %f\nСтандартное отклонение %f""" % (my_rv.mean(), my_rv.var()/5, (my_rv.var()/5)**0.5))
ax = plt.hist(avgs_for_5, bins=20, normed=True)
x = np.linspace(-0.5, 3, 90)
plt.plot(x, norm_for_5.pdf(x))
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
None
```

```python
samples_n10 = [my_rv.rvs(size=10) for i in range(mnogo)]
avgs_for_10 = [sum(rv)/len(rv) for rv in samples_n10]
norm_for_10 = sts.norm(loc=my_rv.mean(), scale=(my_rv.var()/10)**0.5)

print("""Параметры для нормального распределения средних значений случайной выборки 10 Среднее значение %f\nДисперсия %f\nСтандартное отклонение %f""" % (my_rv.mean(), my_rv.var()/10, (my_rv.var()/10)**0.5))

ax = plt.hist(avgs_for_10,bins=20,normed=True)
x = np.linspace(0,4,90)
plt.plot(x, norm_for_10.pdf(x))
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
None
```

С увеличением размера случайной выборки разброс средних значений выборки уменьшается.

следовательно Центральная предельная теорема работает и для Экспоненциального распределения
