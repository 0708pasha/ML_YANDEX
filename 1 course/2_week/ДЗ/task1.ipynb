```python
import re
import pandas as pd
import numpy as np
from scipy.spatial import distance
```

```python
l = []
with open ('sentences.txt', 'r') as file_obj:
    for line in file_obj:
        k = line.lower().strip()
        t = re.split('[^a-z]',k)
        t = [x for x in t if x]
        l += [t]
```

```python
ind = 0
d = dict()
s = set()
for i in l:
    for j in i:
        s.add(j)
for i in s:
    d[ind] = i
    ind += 1
```

```python
m = np.zeros(shape =(len(l),len(d)))
```

```python
for i in range(len(l)):
    for key in d:
        n = l[i].count(d[key])
        m[i,key] = n
```

```python
cos = []
for i in range(len(l)):
    n = distance.cosine(m[0,],m[i,])
    cos.append(n)
cos
```

```python
min1 = min2 = cos[1]
for i in cos[1:]:
    if i <= min1:
        min1,min2 = i, min1
a = cos.index(min2)
b = cos.index(min1)
print(a,b)
```
