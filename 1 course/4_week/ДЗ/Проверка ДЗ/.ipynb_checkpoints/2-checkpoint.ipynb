# Центральная предельная теорема своими руками

```python
# Импорт библиотек для необходимых вычислений, генерации выборок и построения графиков
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sts
% matplotlib inline
```

В качестве непрерывного распределения взято распределение Lomax (частный вид распределения Парето) со значением параметра c = 10. Подробнее о распределении можно узнать в Википедии: https://en.wikipedia.org/wiki/Lomax_distribution описание распределения Lomax в модуле Scipy.stats: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.lomax.html

```python
# Генерируем из распределения Lomax выборку объёма 1000 наблюдений
c = 10.
LX_rv = sts.lomax(c)
sample = LX_rv.rvs(size = 1000)
```

```python
# Построение гистограммы выборки и графика теоретической плотности распределения Lomax
x = np.linspace(0,1, 10)
pdf = LX_rv.pdf(x)
plt.hist(sample, density = True, label = 'Sample')
plt.plot(x, pdf, label = 'Lomax theoretical PDF')
plt.title('Lomax: Sample vs. theoretical PDF')
plt.xlabel('$x$')
plt.ylabel('$F(x)$')
plt.legend()
plt.show()
```

При оценке распределения выборочного среднего Lomax при разных объёмах выборок (5, 10, 50 и 100 наблюдений) были получены следующие результаты:

```python
# Для каждого значения n генерируем 1000 выборок с распределением Lomax, берем выборочные средние
for n in [5, 10, 50, 100]:
    sample_means = []
    for i in range (1000):
        sample = LX_rv.rvs(n)
        sample_means.append(sample.mean())
    x = np.linspace(0, 0.4, 1000)
# Вычисление параметров нормального распределения, аппроксимирующего распределние Lomax с параметром c=10 
    mu = 1/(c-1)
    sigma = np.sqrt(c/( (c-1)**2 * (c-2)) / n)
    norm_rv = sts.norm(loc = mu , scale = sigma)
    pdf = norm_rv.pdf(x)
# Построение совмещенного графика выборочных средних и нормального распределения
    plt.hist(sample_means, bins = 25, density = True, label = 'Sample averages')
    plt.plot(x, pdf, label = 'Normal theoretical PDF')
    plt.title('Lomax: Samples averages vs. normal theoretical PDF N =%d'% n)
    plt.xlabel('Sample average / $x$')
    plt.ylabel('Frequency / $F(x)$')
    plt.legend()
    plt.show()
```
