```python
import numpy as np
import math
```

```python
def f(x):
    return math.sin(x/5.0)*math.exp(x/10.0)+5*math.exp(-x/2.0)
```

```python
f(0)
```

```python
A = np.array([[1.,1.],[1.,15.]])
b = np.array([f(1.),f(15.)])
```

```python
r2 = np.linalg.solve(A,b)
```

```python
r2
```

```python
def f2(x):
    return r2[0]+r2[1]*x
```
