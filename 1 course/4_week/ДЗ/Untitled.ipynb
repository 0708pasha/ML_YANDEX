```python
summ = 1334000.0
vznos = 346840.0
srok = 60
stavka = 0.16
ost = 0
mes = (summ - vznos - summ*ost) * (stavka/12) * (1 + stavka/12)**srok / ((1+stavka/12)**srok-1)
pere = (vznos + mes*srok + summ*ost) - summ 
ostat = summ*ost
```

```python
print('Первоначальный взнос: ' +  str(round(vznos)))
print('Ежемесячный платеж: ' +  str(round(mes)))
print('Переплата: ' +  str(round(pere)))
print('Остаточный платеж: ' +  str(round(ost)))
```

```python
stavka1 = 0.099
ost1 = [0.21,0.30,0.35,0.37,0.40,0.45,0.50]
for i in ost1:
    mes1 = (summ - vznos - i*summ) * (stavka1/12) * (1 + stavka1/12)**srok / ((1+stavka1/12)**srok-1)
    pere1 = (vznos + mes1*srok + summ*i) - summ 
    ostat1 = summ*i
    print('Первоначальный взнос: ' +  str(round(vznos)))
    print('Остаточный платеж: ' +  str(round(ostat1)))
    print('Ежемесячный платеж: ' +  str(round(mes1)))
    print('Переплата: ' +  str(round(pere1)))
    print('')
```
